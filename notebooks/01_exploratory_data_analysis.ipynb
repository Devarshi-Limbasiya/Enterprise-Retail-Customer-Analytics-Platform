```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('master_dataset.csv')
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].dt.year

sales = df.groupby('year')['sales_amount'].sum()

plt.figure(figsize=(10,5))

sales.plot(kind='bar')

plt.title(
    'Annual Sales Revenue',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Year',
    fontsize=12
)

plt.ylabel(
    'Total Sales Revenue ($)',
    fontsize=12
)

plt.xticks(rotation=45)

plt.grid(
    axis='y',
    alpha=0.3
)

plt.tight_layout()

plt.show()
```

```python
country = (
    df.groupby('country')['sales_amount']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))

country.plot(kind='bar')

plt.title(
    'Top Countries by Sales Revenue',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Country',
    fontsize=12
)

plt.ylabel(
    'Total Sales Revenue ($)',
    fontsize=12
)

plt.xticks(rotation=90)

plt.grid(
    axis='y',
    alpha=0.3
)

plt.tight_layout()

plt.show()
```
