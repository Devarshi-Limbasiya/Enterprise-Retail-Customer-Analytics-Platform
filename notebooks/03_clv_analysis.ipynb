```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clv = pd.read_csv('clv_dataset.csv')
clv.head()
```

```python
clv.sort_values(
    'Customer_Lifetime_Value',
    ascending=False
).head(10)
```

```python
import matplotlib.pyplot as plt

top = clv.sort_values(
    'Customer_Lifetime_Value',
    ascending=False
).head(20)

plt.figure(figsize=(12,5))

plt.bar(
    top['customer_key'],
    top['Customer_Lifetime_Value']
)

plt.title(
    'Top 20 Customers by Customer Lifetime Value',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Customer ID',
    fontsize=12
)

plt.ylabel(
    'Customer Lifetime Value ($)',
    fontsize=12
)

plt.xticks(rotation=90)

plt.grid(
    axis='y',
    alpha=0.3
)

plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(clv['Customer_Lifetime_Value'])

plt.title('Customer Lifetime Value Distribution')
plt.xlabel('Customer Lifetime Value ($)')
plt.ylabel('Number of Customers')

plt.grid(alpha=0.3)
plt.show()
```
