```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cohort = pd.read_csv(
    'cohort_retention.csv',
    index_col=0
)

cohort.head()
```

```python
plt.figure(figsize=(15,8))

sns.heatmap(
    cohort,
    cmap='Blues',
    annot=False
)

plt.title(
    'Customer Retention Cohort Analysis',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Months Since First Purchase',
    fontsize=12
)

plt.ylabel(
    'Customer Cohort (Year-Month)',
    fontsize=12
)

plt.tight_layout()

plt.show()
```
