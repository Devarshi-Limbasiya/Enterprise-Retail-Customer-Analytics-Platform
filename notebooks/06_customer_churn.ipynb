```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
importance = {
    'Recency':0.96,
    'Monetary':0.02,
    'Frequency':0.02
}

importance
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    importance.keys(),
    importance.values()
)

plt.title(
    'Feature Importance for Customer Churn Prediction',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Customer Features',
    fontsize=12
)

plt.ylabel(
    'Feature Importance Score',
    fontsize=12
)

plt.grid(
    axis='y',
    alpha=0.3
)

plt.tight_layout()

plt.show()
```
