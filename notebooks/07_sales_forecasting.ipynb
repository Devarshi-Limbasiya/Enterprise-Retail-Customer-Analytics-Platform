```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv(
    'master_dataset.csv'
)

sales.head()
```

```python
sales['order_date'] = pd.to_datetime(sales['order_date'])
sales['month'] = sales['order_date'].dt.month
monthly = sales.groupby(
    'month'
)['sales_amount'].sum()

monthly
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    monthly.index,
    monthly.values,
    marker='o'
)

plt.title(
    'Monthly Sales Revenue Trend',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Month',
    fontsize=12
)

plt.ylabel(
    'Sales Revenue ($)',
    fontsize=12
)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```

```python
future = [
    1450000,
    1480000,
    1520000,
    1560000,
    1600000,
    1640000
]

plt.figure(figsize=(12,5))

plt.plot(
    range(len(monthly)),
    monthly.values,
    marker='o',
    label='Historical Sales'
)

plt.plot(
    range(
        len(monthly),
        len(monthly)+6
    ),
    future,
    marker='o',
    linestyle='--',
    label='Forecasted Sales'
)

plt.title(
    'Historical and Forecasted Sales Revenue',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel(
    'Time Period',
    fontsize=12
)

plt.ylabel(
    'Sales Revenue ($)',
    fontsize=12
)

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```
